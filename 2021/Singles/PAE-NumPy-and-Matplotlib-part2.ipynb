## Plot function

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1,1,20)
y = x**2
plt.plot(x,y)
```

### Multiple graphs

```python
line_2_3 = plt.plot(x,y,x, x**4)

plt.setp(line_2_3,color='red',linewidth = 2, linestyle = '--')

line_4 = plt.plot(x, x**3)
plt.setp(line_4, color='orange', linewidth = 5, linestyle = ':')
```

## Adding text to a graph

```python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 10.0, 1.0
x = np.random.normal(mu, sigma, size=10000)

# gera um histograma de x
plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Hello Boss')

plt.text(6.0, 0.35, 'mu=10, sigma=1.0')
plt.axis([5, 15, 0, 0.5])

plt.grid(True)
```

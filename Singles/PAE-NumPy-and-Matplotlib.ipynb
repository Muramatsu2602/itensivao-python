A estrutura de dados base do *numpy* sao os **arrays**

```python
import numpy as np

# criando um array unidimensional a partir de uma lista
lst = [1,3,5,7,9,10]
a1d = np.array(lst)
print(a1d)

print(lst)
```

```python
b1d = np.zeros((8))
print('ald=', b1d)

b1d = np.ones((8))
print('bld=', b1d)

b1d = np.arange((8))
print('cld=', b1d)

b1d = np.linspace(1,2,5)
print('dld=', b1d)
```
